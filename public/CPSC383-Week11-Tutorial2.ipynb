```python
#Association Rule Mining
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
dataset = [["item1","item2","item5"],
                         ["item2","item4"],
                         ["item2","item3"],
                         ["item1","item2","item4"],
                         ["item1","item3"],
                         ["item2","item3"],
                         ["item1","item3"],
                         ["item1","item2","item3","item5"],
                         ["item1","item2","item3"]]
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
display(df)
rule_count = len(dataset)
# Modify here for question 8
threshold = 0.6
frequent_itemsets = apriori(df, min_support=2/rule_count, use_colnames=True)
print(frequent_itemsets)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=threshold, num_itemsets=rule_count)
print(rules)
```
